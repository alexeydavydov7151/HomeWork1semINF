# Задача 2: Дорога в Шереметьево
Авиакомпания обслуживает 50 городов. Известны тарифы на прямые перелёты между городами (некоторые рейсы отсутствуют). Нужно построить матрицу минимальных тарифов — кратчайшие пути между всеми парами городов.

- Рейсы односторонние

- Возможны акции (отрицательные разницы, но без отрицательных циклов)

Формат входных данных
```python
Первая строка: n (число городов, n ≤ 50)
Далее n строк по n чисел: матрица тарифов dist[i][j]
dist[i][j] = стоимость перелёта i→j
dist[i][i] = 0
dist[i][j] = -1 если рейса нет
```

Пример (5 городов):

```python
5
0 10 -1 -1 25
-1 0 15 5 -1
-1 -1 0 8 12
20 -1 -1 0 10
-1 30 -1 -1 0
```
Пример ответа:
```python
0 10 25 15 35
30 0 15 5 25
45 35 0 8 12
20 30 45 0 10
50 40 55 45 0
```

````python
def print_matrix(matrix):
    for row in matrix:
        print(' '.join(str(x) if x < 999999 else '-1' for x in row))

for i, test in enumerate(test_cases, 1):
    print(f"Тест {i}")
    print(f"n = {test['n']}")

    result = your_code(test['n'], test['tariffs'])

    print("Результат:")
    print_matrix(result)
    print("\nОжидаемый:")
    for row in test['expected']:
        print(' '.join(str(x) if x < 999999 else '-1' for x in row))
    print("-" * 60)
````

````python
def find_all_pairs_shortest_paths(n, tariffs):
    INF = 999999  
    dist = [[INF] * n for _ in range(n)]
    
    for i in range(n):
        for j in range(n):
            if i == j:
                dist[i][j] = 0
            elif tariffs[i][j] != -1:
                dist[i][j] = tariffs[i][j]
    # Алгоритм Флойда-Уоршелла
    for k in range(n):  
        for i in range(n): 
            for j in range(n): 
                if dist[i][k] + dist[k][j] < dist[i][j]:
                    dist[i][j] = dist[i][k] + dist[k][j]
    return dist
def print_matrix(matrix):
    for row in matrix:
        print(' '.join(str(x) if x < 999999 else '-1' for x in row))

# Тестовые случаи
test_cases = [
    # Тест 1:
    {
        "n": 5,
        "tariffs": [[0, 10, -1, -1, 25], [-1, 0, 15, 5, -1], [-1, -1, 0, 8, 12], [20, -1, -1, 0, 10], [-1, 30, -1, -1, 0]],
        "expected": [[0, 10, 25, 15, 25], [25, 0, 15, 5, 15], [28, 38, 0, 8, 12], [20, 30, 45, 0, 10], [55, 30, 45, 35, 0]]
    },
    # Тест 2: Полносвязный
    {
        "n": 3,
        "tariffs": [[0,5,9],[16,0,12],[7,3,0]],
        "expected": [[0,5,9],[16,0,12],[7,3,0]]
    },
    # Тест 3: Недостижимые города
    {
        "n": 4,
        "tariffs": [[0,10,-1,-1],[-1,0,20,-1],[-1,-1,0,30],[-1,-1,-1,0]],
        "expected": [[0,10,30,60],[999999,0,20,50],[999999,999999,0,30],[999999,999999,999999,0]]
    }
]

# Запускаем тесты
for i, test in enumerate(test_cases, 1):
    print(f"Тест {i}")
    print(f"n = {test['n']}")
    
    result = find_all_pairs_shortest_paths(test['n'], test['tariffs'])
    
    print("Результат:")
    print_matrix(result)
    
    print("Ожидаемый результат:")
    print_matrix(test['expected'])
    
    # Проверяем корректность
    correct = True
    for x in range(test['n']):
        for y in range(test['n']):
            if result[x][y] != test['expected'][x][y]:
                correct = False
                break
    
    print(f"Тест пройден: {correct}")
    print("-" * 50)
````
